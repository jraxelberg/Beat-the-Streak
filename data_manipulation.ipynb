```python
hit_list = ['single', 'double', 'triple', 'home_run']
k_list = ['strikeout'] #, 'strikeout_double_play']
w_list = ['walk', 'intent_walk', 'hit_by_pitch']
# drop values where there isn't an event ( ignore all pitches that doesn't result in a play)
data_15 = data_15.dropna(subset=['events'])
# add hit boolean
data_15['hits_count'] = data_15['events'].isin(hit_list).astype(int)
data_15['ks_count'] = data_15['events'].isin(k_list).astype(int)
data_15['walks_count'] = data_15['events'].isin(w_list).astype(int)

# df = data_15['events'].unique()
data_15
```

```python
# HPG calculations (hit per game)
h_df = data_15.groupby(['game_pk', 'batter'])['hits_count'].sum().reset_index()

h_df['h_bool'] = h_df['hits_count'].apply(lambda x: 0 if x == 0 else 1)
h_df['cumulative_game_hit'] = h_df.groupby('batter')['h_bool'].cumsum()
h_df['games_played'] = h_df.groupby('batter').cumcount() + 1

h_df['hpg/year'] = h_df['cumulative_game_hit']/h_df['games_played']

# df['last_5_sum'] = df['value'].rolling(window=5, min_periods=1).sum()

h_df['h_last_10_games'] = h_df.groupby('batter')['h_bool'].rolling(window = 10, min_periods = 1).sum().reset_index(level=0, drop=True)
h_df['h_last_5_games'] = h_df.groupby('batter')['h_bool'].rolling(window = 5, min_periods = 1).sum().reset_index(level=0, drop=True)
h_df['h_last_3_games'] = h_df.groupby('batter')['h_bool'].rolling(window = 3, min_periods = 1).sum().reset_index(level=0, drop=True)
```

```python
# KPG calulation (k's or strikeouts per game)
k_df = data_15.groupby(['game_pk', 'batter'])['ks_count'].sum().reset_index()
k_df['cumulative_ks'] = k_df.groupby('batter')['ks_count'].cumsum()
k_df['k_last_10_games'] = k_df.groupby('batter')['ks_count'].rolling(window = 10, min_periods = 1).sum().reset_index(level=0, drop=True)
k_df['k_last_5_games'] = k_df.groupby('batter')['ks_count'].rolling(window = 5, min_periods = 1).sum().reset_index(level=0, drop=True)
k_df['k_last_3_games'] = k_df.groupby('batter')['ks_count'].rolling(window = 3, min_periods = 1).sum().reset_index(level=0, drop=True)


# WPG calculations (walks per game)
w_df = data_15.groupby(['game_pk', 'batter'])['walks_count'].sum().reset_index()
w_df['cumulative_ws'] = w_df.groupby('batter')['walks_count'].cumsum()
w_df['w_last_10_games'] = w_df.groupby('batter')['walks_count'].rolling(window = 10, min_periods = 1).sum().reset_index(level=0, drop=True)
w_df['w_last_5_games'] = w_df.groupby('batter')['walks_count'].rolling(window = 5, min_periods = 1).sum().reset_index(level=0, drop=True)
w_df['w_last_3_games'] = w_df.groupby('batter')['walks_count'].rolling(window = 3, min_periods = 1).sum().reset_index(level=0, drop=True)

# ABS calculations (at bats per game)
ab_df = data_15.groupby(['game_pk', 'batter'])['ab_count'].sum().reset_index()
ab_df['cumulative_ws'] = ab_df.groupby('batter')['ab_count'].cumsum()
ab_df['ab_last_10_games'] = ab_df.groupby('batter')['ab_count'].rolling(window = 10, min_periods = 1).sum().reset_index(level=0, drop=True)
ab_df['ab_last_5_games'] = ab_df.groupby('batter')['ab_count'].rolling(window = 5, min_periods = 1).sum().reset_index(level=0, drop=True)
ab_df['ab_last_3_games'] = ab_df.groupby('batter')['ab_count'].rolling(window = 3, min_periods = 1).sum().reset_index(level=0, drop=True)

join_1 =  pd.merge(h_df, k_df, on=['game_pk', 'batter'], how='left')
join_2 =  pd.merge(join_1, w_df, on=['game_pk', 'batter'], how='left')
join_3 = pd.merge(join_2, ab_df, on=['game_pk', 'batter'], how='left')

join_3['kpg/year'] = join_3['cumulative_ks']/join_3['games_played']
join_3['wpg/year'] = join_3['cumulative_ws']/join_3['games_played']
join_3['ab/year'] = join_3['cumulative_ws']/join_3['games_played']


join_3
```

```python
all_data =  pd.merge(join_2, data_15, on=['game_pk', 'batter'], how='left')
all_data
```
